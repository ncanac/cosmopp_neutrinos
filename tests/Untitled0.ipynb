```python
from bao_boss_like import bao_boss
```

```python
params = {
    'output': 'tCl lCl',
    'l_max_scalars': 2,
    'lensing': 'yes',
    'A_s': 2.2154e-9,
    'n_s': 0.9619, 
    'h': 0.6704,
    'omega_b': 0.022032,
    'omega_cdm': 0.12038}
```

```python
like = bao_boss('bao_2014.txt')
```

```python
like.loglkl(params)
```

```python
from classy import Class
```

```python
cosmo = Class()
cosmo.set(params)
cosmo.compute()
```

```python
cosmo.rs_drag()
```

```python
for z in like.z:
    da = cosmo.angular_distance(z)
    dr = z/cosmo.Hubble(z)
    dv = pow(da*da*(1+z)*(1+z)*dr,1./3.)
    rs = cosmo.rs_drag()
    print da, dr, dv, rs
```
